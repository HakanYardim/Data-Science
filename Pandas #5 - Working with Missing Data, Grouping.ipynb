# groupby

```python
wages = {"Departments": ["Software","Hardware",
"Human Resources", "Project Management","Finance","Law"],
"Workers":["Cagan","Poyraz","Hakan","Mehmet","Vedat","Emre"],
"Maas": [100,130,224,231,312,122]}
```

```python
wagesDataFrame = pd.DataFrame(wages)
```

```python
wagesDataFrame
```

```python
departments = wagesDataFrame.groupby("Departments")
```

```python
departments.count()
```

```python
departments.mean()
```

```python
departments.min()
```

```python
departments.max()
```

```python
departments.describe()
```
